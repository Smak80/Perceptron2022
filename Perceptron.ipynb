```python
import loader as l
ld = l.loader()
tri = ld.getTrainInp()
tro = ld.getTrainOut()
tsi = ld.getTestInp()
tso = ld.getTestOut()
from MLPerceptron import MLP
p = MLP(tri, tro, tsi, tso, neuronNum=(10, 7))
err_tr, err_ts = p.learn(eta=0.025, epsilon=25e-4, epoches=3000)
```

```python
from matplotlib import pyplot as plt
f1 = plt.figure(1)
fa1 = f1.add_subplot(1, 1, 1)
out = p.calc(tri)
fa1.plot(tri, out, "bo")
fa1.plot(tri, tro, "r+")
out = p.calc(tsi)
fa1.plot(tsi, out, "gv")
fa1.plot(tsi, tso, "y+")
f2 = plt.figure(2)
err_tr_x = [i for i in range(1, len(err_tr)+1)]
err_ts_x = [i for i in range(1, len(err_ts)+1)]
fa2 = f2.add_subplot(2, 1, 1)
fa2.plot(err_tr_x, err_tr, "r-")
fa3 = f2.add_subplot(2, 1, 2)
fa3.plot(err_ts_x, err_ts, "b-")
plt.show()
```
